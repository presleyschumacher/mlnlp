```python
import numpy as np
import matplotlib.pyplot as plt

data = {'a': 62, 't': 83, ' ': 181, 'h': 61, 'r': 45, 'e': 105, 'o': 53, '’': 2, 'c': 25, 'l': 26, 'k': 6, 'p': 6, 'i': 72, 's': 55, 'y': 18, 'w': 21, 'b': 10, ',': 15, 'u': 17, 'm': 20, 'd': 36, 'n': 49, '.': 6, 'f': 19, 'g': 15, 'v': 8, 'q': 2, 'x': 2}
letters = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))

plt.bar(letters, values, color ='blue', width = 0.4, )
plt.xlabel("Letters")
plt.ylabel("Frequency")
plt.title("Frequency Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

sizes = [10, 20, 50, 100, 1000, 5000]
insertion_sort_times = [0.0010252999999999998, 0.0027026999999999954, 0.010147200000000002, 0.0381137, 3.6303399, 91.2180796]
merge_sort_times = [0.00161889999999687, 0.003635600000009731, 0.0090655000000055, 0.020108000000007564, 0.2687474000000094, 1.6147050999999948]
                    
x1 = sizes
y1 = insertion_sort_times
plt.plot(x1, y1, marker='o', color = 'blue', label = "insertion_sort_times")

x1 = sizes
y2 = merge_sort_times
plt.plot(x1, y2, marker='o', color = 'green', label = "merge_sort_times")

plt.xlabel('Time')
plt.ylabel('Sizes')
plt.title('Time to Sort an Algorithm based on Size')
plt.legend()

plt.show()
```
